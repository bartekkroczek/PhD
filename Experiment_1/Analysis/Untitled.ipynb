```python
import numpy as np
import matplotlib.pyplot as plt

def analyze_fir_filter(b, fs, n_points=1024):
    """
    Analyze FIR filter characteristics with real frequencies.
    
    Parameters:
    -----------
    b : array_like
        FIR filter coefficients
    fs : float
        Sampling frequency in Hz
    n_points : int, optional
        Number of points for FFT computation (default=1024)
        
    Returns:
    --------
    dict
        Dictionary containing filter characteristics and analysis results
    """
    # Calculate frequency response
    w = np.fft.rfftfreq(n_points, d=1/fs)  # This gives frequencies in Hz
    h = np.fft.rfft(b, n=n_points)
    
    # Calculate magnitude and phase responses
    magnitude = np.abs(h)
    phase = np.unwrap(np.angle(h))
    
    # Calculate key metrics
    max_magnitude = np.max(magnitude)
    max_freq = w[np.argmax(magnitude)]
    
    # Calculate -3dB points (if they exist)
    threshold = max_magnitude / np.sqrt(2)
    cutoff_frequencies = []
    for i in range(1, len(magnitude)):
        if (magnitude[i-1] - threshold) * (magnitude[i] - threshold) <= 0:
            # Linear interpolation to find more precise cutoff frequency
            freq_ratio = (threshold - magnitude[i-1]) / (magnitude[i] - magnitude[i-1])
            cutoff_freq = w[i-1] + freq_ratio * (w[i] - w[i-1])
            cutoff_frequencies.append(cutoff_freq)
    
    # Calculate group delay (in seconds)
    group_delay = -np.diff(phase) / np.diff(w * 2 * np.pi)  # Convert to seconds
    
    # Set APA style parameters
    plt.style.use('default')
    plt.rcParams['font.family'] = 'Times New Roman'
    plt.rcParams['font.size'] = 12
    plt.rcParams['axes.titlesize'] = 12
    plt.rcParams['axes.labelsize'] = 12
    plt.rcParams['xtick.labelsize'] = 10
    plt.rcParams['ytick.labelsize'] = 10
    plt.rcParams['legend.fontsize'] = 10
    plt.rcParams['figure.titlesize'] = 12

    
    # Create visualization with APA style
    fig, ax1 = plt.subplots(1, 1, figsize=(8, 4))  # APA typically prefers smaller figures
    
    # Create ideal low-pass filter response
    ideal_response = np.ones_like(w)
    ideal_response[w > 15] = 0  # Full attenuation after 15 Hz
    
    # Magnitude response
    ax1.plot(w, magnitude, color='black', linewidth=1.5, label='Actual Response')  # APA prefers black lines
    ax1.plot(w, ideal_response, 'r--', linewidth=1.5, label='Ideal Low-pass (15 Hz)', alpha=0.7)
    ax1.set_title('Frequency Response Magnitude plot of an moving average filter \n compared to idealised Low-Pass filter', pad=10)
    ax1.set_xlabel('Frequency (Hz)', labelpad=10)
    ax1.set_ylabel('Magnitude', labelpad=10)
    
    if cutoff_frequencies:
        ax1.axhline(y=threshold, color='black', linestyle=':', label='-3dB line', linewidth=0.8)
        for f in cutoff_frequencies:
            ax1.axvline(x=f, color='black', linestyle=':', label=f'-3dB at {f:.1f} Hz', linewidth=0.8)
    
    # APA style legend
    ax1.legend(frameon=True, edgecolor='black', fancybox=False)
    
    # Adjust layout
    plt.tight_layout()
    
    # Add box around the plot (APA requirement)
    #for spine in ax1.spines.values():
        #spine.set_linewidth(0.5)
        #spine.set_visible(False)
        
    plt.gca().spines['top'].set_visible(False)
    plt.gca().spines['right'].set_visible(False)
    plt.gca().spines['bottom'].set_linewidth(0.5)
    plt.gca().spines['left'].set_linewidth(0.5)
    
    # Prepare analysis results
    results = {
        'max_magnitude': max_magnitude,
        'max_freq': max_freq,
        'cutoff_frequencies': cutoff_frequencies,
        'filter_length': len(b),
        'average_group_delay': np.mean(group_delay) * 1000,  # in ms
        'frequency_response': h,
        'frequencies': w,
        'magnitude_response': magnitude,
        'phase_response': phase,
        'group_delay': group_delay * 1000,  # in ms
        'sampling_frequency': fs
    }
    
    return results, fig

def print_filter_analysis(results):
    """Print a formatted analysis of the filter characteristics."""
    print("\nFIR Filter Analysis:")
    print("-" * 50)
    print(f"Sampling Frequency: {results['sampling_frequency']} Hz")
    print(f"Filter Length: {results['filter_length']} taps")
    print(f"Maximum Magnitude: {results['max_magnitude']:.4f}")
    print(f"Frequency at Maximum Magnitude: {results['max_freq']:.1f} Hz")
    print(f"Average Group Delay: {results['average_group_delay']:.2f} ms")
    
    if results['cutoff_frequencies']:
        print("\nCutoff Frequencies (-3dB points):")
        for i, freq in enumerate(results['cutoff_frequencies']):
            print(f"  f{i+1} = {freq:.1f} Hz")
    else:
        print("\nNo -3dB cutoff frequencies found")
```
